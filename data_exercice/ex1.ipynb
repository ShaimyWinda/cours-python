```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('house_pricing.csv')
data.describe()
# info : data.info()
```

```python
#Liste des types
data.dtypes
```

```python
#Nuage de point LotArea par rapport au SalePrice
sns.relplot(x="SalePrice", y="LotArea", data=data)
```

```python
# Nuage de point LotArea par rapport SalePrice 
# avec un SalePrice < 500 000 et LotArea < 20 000
# Correction :
# filter_lot_area_and_price = data[data.LotArea<20000][data.SalePrice<500000]
# sns.relplot(x='LotArea', y='SalePrice', data=filter_lot_area_and_price)
sns.relplot(x="SalePrice", y="LotArea", data=data)
plt.xlim(0, 500000)
plt.ylim(0, 20000)
```

```python
# Nuage de point LotFrontage par rapport LotArea
sns.relplot(x="LotArea", y="LotFrontage", data=data)
```

```python
# Nuage de point
# LotArea < 100 000 et LotFrontage < 200
sns.relplot(x="LotArea", y="LotFrontage", data=data)
plt.xlim(0, 100000)
plt.ylim(0, 200)
```
