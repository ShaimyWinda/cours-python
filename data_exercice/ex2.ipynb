```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
myData = pd.read_csv('sales_predictions.csv')
```

```python
sns.relplot(x="date", y="item_price", kind="line", data=myData)
# Resultat pas clair
```

```python
# graphique sur 50 donnees
sns.relplot(x="date", y="item_price", kind="line", data=myData.sample(50))
```

```python
f, ax = plt.subplots(figsize=(20,6))
sns.relplot(x="date", y="item_price", data=myData.sample(50), kind="line", ax = ax)
plt.show()
```

```python
date =  pd.to_datetime(myData['date'])
sns.relplot(x=date, y="item_price", data = myData.sample(50), kind="line", ax = ax)
plt.show()
```
